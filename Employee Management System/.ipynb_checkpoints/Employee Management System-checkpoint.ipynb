```python
import csv

def display_menu():
    """Displays the main menu"""
    print("Employee Management System")
    print("1. Add New Employee")
    print("2. Edit Existing Employee")
    print("3. Delete Employee")
    print("4. Display All Employees")
    print("5. Exit")

def add_employee():
    """Adds a new employee record"""
    with open('employees.csv', mode='a', newline='') as file:
        writer = csv.writer(file)
        name = input("Enter employee name: ")
        email = input("Enter employee email: ")
        phone = input("Enter employee phone: ")
        writer.writerow([name, email, phone])
    print("Employee added successfully.")

def edit_employee():
    """Edits an existing employee record"""
    name = input("Enter employee name to edit: ")
    with open('employees.csv', mode='r', newline='') as file:
        reader = csv.reader(file)
        employees = list(reader)
        for i in range(len(employees)):
            if employees[i][0] == name:
                email = input("Enter new email: ")
                phone = input("Enter new phone: ")
                employees[i][1] = email
                employees[i][2] = phone
                break
    with open('employees.csv', mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(employees)
    print("Employee record updated successfully.")

def delete_employee():
    """Deletes an existing employee record"""
    name = input("Enter employee name to delete: ")
    with open('employees.csv', mode='r', newline='') as file:
        reader = csv.reader(file)
        employees = list(reader)
        for i in range(len(employees)):
            if employees[i][0] == name:
                employees.pop(i)
                break
    with open('employees.csv', mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(employees)
    print("Employee record deleted successfully.")

def display_employees():
    """Displays a list of all employees"""
    with open('employees.csv', mode='r', newline='') as file:
        reader = csv.reader(file)
        employees = list(reader)
        print("{:<20} {:<30} {:<15}".format("Name", "Email", "Phone"))
        for employee in employees:
            print("{:<20} {:<30} {:<15}".format(employee[0], employee[1], employee[2]))

def main():
    """Main program"""
    while True:
        display_menu()
        choice = input("Enter your choice (1-5): ")
        if choice == '1':
            add_employee()
        elif choice == '2':
            edit_employee()
        elif choice == '3':
            delete_employee()
        elif choice == '4':
            display_employees()
        elif choice == '5':
            print("Thank you for using the Employee Management System!")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == '__main__':
    main()
```
